# Creating Linked List

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next_value = None

        
class SinglyLinkedList:
    
    def __init__(self):
        '''Create Head Node'''
        self.head_value = None
        
    def listPrint(self):
        '''Print entire linked list'''
        print_value = self.head_value
       
        while print_value is not None:
            print (print_value.value)
            print_value = print_value.next_value
        print()
            
    def insertNewHead(self,value):
        '''Inster a New Head Node'''
        new_head = Node(value)
        new_head.next_value = self.head_value
        self.head_value = new_head
        
    def insertNewEnd(self,value):
        '''Inster a New Node at the end of the Linked List'''
        new_node = Node(value)
        if self.head_value is None:
            self.head_value = new_node
            return
        
        last_node = self.head_value
        while last_node.next_value:
            last_node = last_node.next_value
        last_node.next_value = new_node
    
    def insertNode(self,middle_value,value):
        '''Inster a New Node after a middle node'''
        new_node = Node(value)
        
        #if no head value then create it with middle value and link to value
        if self.head_value is None:
            middle_node = Node(middle_value)
            self.head_value = middle_node
            self.head_value.next_value = new_node 
            return
        
        #search through list and find middle value, then link to value 
        next_node = self.head_value
        while next_node:
            if next_node.value == middle_value:
                middle_node = next_node
                new_node.next_value = middle_node.next_value
                middle_node.next_value = new_node
                return 
            else:
                next_node = next_node.next_value
                
        print("Middle Node not Found")
        return
    
        
    def removeNode(self,value):
        '''Remove a Node'''
        
        #If value is head then remove the head node and replace with the next node
        if self.head_value.value == value:
            self.head_value = self.head_value.next_value
            return
        
        #search through list and find value, then remove node and replace next node with previous node
        next_node = self.head_value
        prev_node = self.head_value
        while next_node:

            if next_node.value == value:
                prev_node.next_value = next_node.next_value
                next_node = None
                return 
            else:
                prev_node = next_node
                next_node = next_node.next_value
                
        print("Node not Found")
        return
       

```

```python
linkedList = SinglyLinkedList()

linkedList.insertNewHead("Sun")
linkedList.listPrint()

linkedList.insertNode("Sun","Mon")
linkedList.insertNode("Mon","Tues")
linkedList.insertNode("Tues","Wed")
linkedList.listPrint()

linkedList.removeNode("fdsa")
linkedList.listPrint()

linkedList.insertNewEnd("Wed")
linkedList.listPrint()


linkedList.insertNode("Wed","Thurs")
linkedList.listPrint()
```

```python
linkedList = SinglyLinkedList()

linkedList.head_value = Node("Mon")

# Link first Node to second node
child_node_1 = Node("Tue")
linkedList.head_value.next_value = child_node_1

# Link second Node to third node
child_node_2 = Node("Wed")
child_node_1.next_value = child_node_2


linkedList.insertNewHead("Sun")

linkedList.insertNewEnd("Thurs")

linkedList.insertNewEnd("Sat")

linkedList.insertNode("Thurs","Fri")

linkedList.listPrint()

```
